```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity,euclidean_distances,manhattan_distances
```

```python
cosine_sim = cosine_similarity(data.iloc[[0]],data.iloc[[1]])
ecu_dis = euclidean_distances(data.iloc[[0]],data.iloc[[1]])
man_dis = manhattan_distances(data.iloc[[0]],data.iloc[[1]])
```

```python
print(cosine_sim)
print(ecu_dis)
print(man_dis)
```

```python
from scipy.stats import pearsonr
from sklearn.metrics import jaccard_score
```

```python
corr = pearsonr(data['sepal length (cm)'],data['sepal width (cm)'])
print(corr)
```

```python
data['target'] = iris.target
```

```python
j_c = jaccard_score(data.iloc[0]['target'],data.iloc[1]['target'])
```
