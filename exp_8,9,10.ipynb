```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/IRIS.csv')
```

```python
data
```

```python
x = data.drop('species',axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
kmean_lab=kmeans.predict(x)
```

```python
plt.scatter(x['sepal_length'],x['sepal_width'],c=kmean_lab)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=4,linkage='ward')
ac.fit(x)
ac_lab= ac.labels_
```

```python
plt.scatter(x['sepal_length'],x['sepal_width'],c=ac_lab,cmap='viridis')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
dbscan.fit(x)
dn_lab = dbscan.labels_
```

```python
plt.scatter(x['sepal_length'],x['sepal_width'],c=dn_lab)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```
